```python
# Importing libraries
import pandas as pd
import PyCO2SYS as pyco2
```

```python
# Loading dataset
file_path = "/Users/Mbongeleni/Library/CloudStorage/OneDrive-StellenboschUniversity/Documents/CSIR/SANAE/SANAE_Combined_Underway_clean.xlsx"
df = pd.read_excel(file_path)
```

```python
# Define function to compute carbonate system
def compute_carbonate_system(row):
    results = pyco2.sys(
        par1=row["Total Carbon"],       # DIC (µmol/kg)
        par2=row["Total Alkalinity"],   # TA (µmol/kg)
        par1_type=2,                    # 2 = DIC
        par2_type=1,                    # 1 = TA
        salinity=row["Salinity"],       # Salinity 
        temperature=row["Temperature"], # in-situ temperature (°C)
        pressure=0,                     # surface pressure
        opt_pH_scale=1,                 # total scale
        opt_k_carbonic=14,              # Southern Ocean constants
        opt_k_bisulfate=2               # Southern Ocean constants
    )
    return pd.Series({
        "pCO2": results["pCO2"], 
        "pH": results["pH"], 
        "OmegaAragonite": results["saturation_aragonite"]
    })
```

```python
# Applying function row by row
df[["pCO2", "pH", "OmegaAragonite"]] = df.apply(compute_carbonate_system, axis=1)
df.to_excel(file_path, index=False)
```

```python
# Saving results back to the same file
df.to_excel(file_path, index=False)
print("Carbonate system calculations complete. Columns pCO2, pH, and OmegaAragonite added to the file.")
```
