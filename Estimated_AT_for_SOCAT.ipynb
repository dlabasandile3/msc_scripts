```python
# SOCAT Total Alkalinity Estimation Notebook

import pandas as pd
```

```python
# 1. Load the SOCAT dataset
file_path = "/Users/Mbongeleni/Documents/OneDrive - University of KwaZulu-Natal/Documents/CSIR/SOCAT/SOCAT_combined.xlsx"
df = pd.read_excel(file_path)
```

```python
# 2. Define the Lee et al. (2006) equation for AT
def estimate_at(sss, sst):
    return (
        2305
        + 52.48 * (sss - 35)
        + 2.85 * (sss - 35)**2
        - 0.49 * (sst - 20)
        + 0.086 * (sst - 20)**2
    )
```

```python
# 3. Apply the equation using SAL as SSS and TEMP as SST
df["AT_estimate"] = df.apply(lambda row: estimate_at(row["SAL"], row["TEMP"]), axis=1)
```

```python
# 4. Save the updated dataset back to Excel
df.to_excel(file_path, index=False)

print("AT_estimate column successfully added to SOCAT_combined.xlsx")
```
