```python
# ============================================================
# CONVERT EXCEL FILE (.xlsx) TO NETCDF FILE (.nc)
# ============================================================

# -----------------------------
# 1. IMPORT LIBRARIES
# -----------------------------
import pandas as pd
import xarray as xr

# -----------------------------
```

```python
# 2. INPUT FILE
# -----------------------------
excel_file = "/Users/Mbongeleni/Library/CloudStorage/OneDrive-StellenboschUniversity/Documents/CSIR/CSIR_ML6/CSIR-ML6_filtered.xlsx"
# -----------------------------
```

```python
# 3. READ EXCEL FILE
# -----------------------------
print("Reading Excel file...")

df = pd.read_excel(excel_file)

print("\nPreview of dataframe:")
print(df.head())

print("\nColumns in dataset:")
print(df.columns)

# -----------------------------
```

```python
# 4. OPTIONAL: CONVERT TIME COLUMN
# -----------------------------
# Replace 'time' with your actual time column name if different

if "time" in df.columns:
    df["time"] = pd.to_datetime(df["time"])

# -----------------------------
```

```python
# 5. REMOVE COMPLETELY EMPTY ROWS
# -----------------------------
df = df.dropna(how="all")

# -----------------------------
```

```python
# 6. CONVERT DATAFRAME TO XARRAY
# -----------------------------
print("\nConverting dataframe to xarray dataset...")

# If your file has these coordinate columns:
possible_coords = ["time", "depth", "latitude", "longitude"]

coords = [col for col in possible_coords if col in df.columns]

# Set coordinates if they exist
if len(coords) > 0:
    ds = df.set_index(coords).to_xarray()
else:
    ds = xr.Dataset.from_dataframe(df)

print("\nDataset summary:")
print(ds)

# -----------------------------
```

```python
# 7. OUTPUT NETCDF FILE
# -----------------------------
output_nc = "/Users/Mbongeleni/Library/CloudStorage/OneDrive-StellenboschUniversity/Documents/CSIR/CSIR_ML6/CSIR-ML6_filtered.nc"

# -----------------------------
```

```python
# 8. OPTIONAL COMPRESSION
# -----------------------------
encoding = {
    var: {"zlib": True, "complevel": 5}
    for var in ds.data_vars
}

# -----------------------------
```

```python
# 9. SAVE NETCDF FILE
# -----------------------------
print("\nSaving NetCDF file...")

ds.to_netcdf(
    output_nc,
    encoding=encoding
)

print("\n===================================")
print("CONVERSION COMPLETED SUCCESSFULLY")
print("===================================")

print(f"\nNetCDF file saved to:\n{output_nc}")
```
