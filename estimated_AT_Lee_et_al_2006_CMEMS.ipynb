```python
import xarray as xr
```

```python
# File paths
sst_file = r"C:\Users\dlaba\Downloads\cmems_temperature.nc"
sss_file = r"C:\Users\dlaba\Downloads\cmems_salinity.nc"
```

```python
# Load datasets
sst_ds = xr.open_dataset(sst_file)
sss_ds = xr.open_dataset(sss_file)
```

```python
# Extract variables
sst = sst_ds["thetao"]   # Sea Surface Temperature
sss = sss_ds["so"]       # Sea Surface Salinity

# Apply Lee et al. (2006) equation
AT = (2305
      + 52.48 * (sss - 35)
      + 2.85 * (sss - 35)**2
      - 0.49 * (sst - 20)
      + 0.086 * (sst - 20)**2)

# Drop NaNs
AT = AT.where(~AT.isnull(), drop=True)
```

```python
# Save to new NetCDF file
AT.name = "AT"
AT.to_netcdf(r"C:\Users\dlaba\Downloads\AT.nc")

print("AT.nc file created successfully!")
```
